```python
import pandas as pd
```

```python
df = pd.read_csv("fine fmeasure.csv", sep=";", index_col=0)
df
```

```python
columns = ["ABBR", "DESC", "HUM", "ENTY", "LOC", "NUM"]
anl = pd.DataFrame(columns=columns, index = ["0-0.25", "0.25-0.5", "0.5-0.75", "0.75-1"])
anl
```

```python
for col in columns:
    temp = df[df["Fine Class"].str.startswith(col)]
    
    cnt = len(temp[temp["F-Measure"] <= 0.25])
    anl[col]["0-0.25"] = cnt
    temp = temp[temp["F-Measure"] > 0.25]
    
    cnt = len(temp[temp["F-Measure"] <= 0.5])
    anl[col]["0.25-0.5"] = cnt
    temp = temp[temp["F-Measure"] > 0.5]
    
    cnt = len(temp[temp["F-Measure"] <= 0.75])
    anl[col]["0.5-0.75"] = cnt
    temp = temp[temp["F-Measure"] > 0.75]
    
    cnt = len(temp)
    anl[col]["0.75-1"] = cnt

anl
    
```
