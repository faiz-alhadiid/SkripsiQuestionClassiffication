```python
import pandas as pd
import json
```

```python
c_list = [0.5, 1, 1.5, 2, 2.5, 3]
max_iter_list = [250, 500, 750, 1000, 1500, 2000, 5000]
tol_list = [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5]
```

```python
# Pengujian C

max_iter = 500
tol = 0.001

df = pd.DataFrame(columns=['C','Fold', 'Accuracy', 'Precision', 'Recall', 'F Measure'])
for c in c_list:
    for i in range(5):
        file_name = f"fine_test/C_{c}-tol_{tol}-max_iter_{max_iter}-fold_{i}.json"
        f = open(file_name, 'r')
        json_file = json.loads(f.read())
        evaluation = json_file['evaluation']
        df = df.append({
            'C': c, 
            'Fold': i+1, 
            'Accuracy': round(evaluation['accuracy'], 5),
            'Precision': round(evaluation['avg_precision'], 5),
            'Recall': round(evaluation['avg_recall'], 5),
            'F Measure': round(evaluation['avg_f_measure'], 5)
        }, ignore_index=True)


df.to_csv('csv_fine/pengujian_c.csv', sep=';', index=False)
df
        
        
```
